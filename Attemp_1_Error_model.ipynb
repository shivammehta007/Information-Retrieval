#### We have all words in the words dataframe

```python
distance_to_keep = [1, 2]
distance_function = textdistance.levenshtein
```

```python
textdistance.levenshtein.distance('abc', 'abc')
```

```python
# Computing Distance for each word from the dataset keeping distance 1 or 2
# Change this to other distances to check for other distances

def check_spelling(word):
    nearby_words = []
    for w in words:
        distance = distance_function.distance(w, word)
        if distance == 0:
            return w
        if distance in distance_to_keep:
            nearby_words.append((w, words[w]))
    return sorted(nearby_words, key=lambda x: x[1], reverse=True)[0][0]
    
    
```

```python
start_time = time.time()
print(check_spelling('ПОРНОСПЯЩИЕ'))
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
test_data['Predicted'] = test_data['Id'].apply(check_spelling)
```
